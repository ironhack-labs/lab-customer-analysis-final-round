```python
#finding outliers using Quartile 
'''
A data point is an outlier if it is more than 1.5 ⋅ IQR ‍ above the third quartile or below the first quartile. 
Said differently, low outliers are below Q 1 − 1.5 ⋅ IQR ‍ and high outliers are above Q 3 + 1.5 ⋅ IQR ‍ .
'''

def find_outliers_IQR(df):

    q1=df.quantile(0.25)
    q3=df.quantile(0.75)
    IQR=q3-q1
    outliers = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]
    return outliers

numerical_na =find_outliers_IQR(h_corr)
```

```python
outliers = find_outliers_IQR(h_corr)

print("number of outliers: "+ str(len(outliers)))

print("max outlier value: "+ str(outliers.max()))

print("min outlier value: "+ str(outliers.min()))

#income is well distributed
```

```python
def drop_outliers_IQR(df):
    
    q1=df.quantile(0.25)
    q3=df.quantile(0.75)
    IQR=q3-q1
    not_outliers = df[~((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]

    outliers_dropped = not_outliers.dropna()

    return outliers_dropped

no_outlier = drop_outliers_IQR(h_corr)
no_outlier
```
