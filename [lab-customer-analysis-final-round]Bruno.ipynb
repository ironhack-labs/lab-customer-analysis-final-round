```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
import statsmodels.api as sm
from statsmodels.formula.api import ols
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
marketing_costumer = pd.read_csv(r"C:\Users\Utilizador\Desktop\Ironhack 2\LabRoundFinal\lab-customer-analysis-final-round\files_for_lab\csv_files\marketing_customer_analysis.csv")
marketing_costumer
```

```python
cols = []
for i in range(len(marketing_costumer.columns)):
    cols.append(marketing_costumer.columns[i].lower().replace(' ', '_')) 
marketing_costumer.columns = cols
```

```python
marketing_costumer
```

```python
marketing_costumer.isna().sum()
```

```python
marketing_costumer.drop_duplicates()
```

```python
numericals = marketing_costumer.select_dtypes(include=['float64', 'int64'])
numericals
```

```python
categoricals = marketing_costumer.select_dtypes('object')
categoricals
```

```python
numericals.drop(['months_since_last_claim', 'months_since_policy_inception','number_of_open_complaints', 'number_of_policies'], axis=1,inplace=True)
```

```python
numericals.columns
```

```python
categoricals.drop(['customer','effective_to_date', 'response', 'vehicle_size','policy' ],axis = 1, inplace = True)
```

```python
categoricals.columns
```

```python
numericals['income'] = numericals['income'].replace(0, numericals['income'].mean())
```

```python
for col in categoricals.columns:
    categoricals[col]=LabelEncoder().fit(categoricals[col]).transform(categoricals[col])
    
    #This code is transforming categorical data in each column of the "categoricals" DataFrame into numerical format using label encoding. It replaces categorical values with unique numerical labels
```

```python
marketing_costumer = pd.concat([numericals, categoricals], axis=1)
```

```python
marketing_costumer
```

```python
X = marketing_costumer.drop(['total_claim_amount'],axis = 1)
Y = marketing_costumer[['customer_lifetime_value','total_claim_amount', 'income', 'monthly_premium_auto','state','education']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.8, random_state=140)
```

```python
lm = linear_model.LinearRegression()
model = lm.fit(X_train,y_train)

model.score(X_train,y_train)  


predictions = model.predict(X_test)  
predictions

print("R2 value is =",(r2_score(y_test,predictions),))
print ("The intercept of the model is =",lm.intercept_)
print ("the coefficients of the model are =", lm.coef_)
from sklearn.metrics import mean_absolute_error
mse = mean_squared_error(y_test, predictions)
mae = mean_absolute_error(y_test, predictions)
print("The mse of the model is = ",mse)
print("The root mse of the model is = ",np.sqrt(mse))
print("The mean absolute error of the model is = ",mae)
```
